### Import Libraries and Setup Kaggle

```python
# Import required libraries
import os
import pandas as pd
import numpy as np
import kagglehub
from pathlib import Path

# Set up paths
PROJECT_ROOT = Path.cwd().parent  # Go up one level from notebooks folder
ARTIFACTS_DIR = PROJECT_ROOT / 'artifacts'
RAW_DATA_DIR = ARTIFACTS_DIR / 'raw'
PROCESSED_DATA_DIR = ARTIFACTS_DIR / 'processed'

# Create directories if they don't exist
RAW_DATA_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)

print(f"Project root: {PROJECT_ROOT}")
print(f"Raw data dir: {RAW_DATA_DIR}")
print(f"Processed data dir: {PROCESSED_DATA_DIR}")
```

### Download Dataset using KaggleHub

```python
# Download dataset from Kaggle
print("Downloading dataset from Kaggle...")
path = kagglehub.dataset_download("sriharshaeedala/financial-fraud-detection-dataset")
print(f"Dataset downloaded to: {path}")

# List files in the downloaded directory
print("\nFiles in dataset:")
for file in os.listdir(path):
    print(f"  - {file}")
    file_path = os.path.join(path, file)
    file_size = os.path.getsize(file_path) / (1024 * 1024)  # Size in MB
    print(f"    Size: {file_size:.2f} MB")
```

### Load the Dataset

```python
# Find and load the CSV file
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]

if csv_files:
    dataset_path = os.path.join(path, csv_files[0])
    print(f"Loading: {csv_files[0]}")
    df = pd.read_csv(dataset_path)
    
    # Save a copy to artifacts/raw folder for backup
    raw_copy_path = RAW_DATA_DIR / csv_files[0]
    df.to_csv(raw_copy_path, index=False)
    print(f"Saved raw copy to: {raw_copy_path}")
else:
    print("No CSV file found in the downloaded dataset")
```

### Initial Data Exploration

```python
# Basic dataset information
print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)

print(f"\nDataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"\nMemory usage: {df.memory_usage(deep=True).sum() / (1024**2):.2f} MB")

print("\n" + "=" * 60)
print("FIRST 5 ROWS")
print("=" * 60)
display(df.head())

print("\n" + "=" * 60)
print("LAST 5 ROWS")
print("=" * 60)
display(df.tail())

print("\n" + "=" * 60)
print("COLUMN INFORMATION")
print("=" * 60)

# Column info
column_info = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Non-Null Count': df.count().values,
    'Null Count': df.isnull().sum().values,
    'Null %': (df.isnull().sum() / len(df) * 100).values,
    'Unique Values': df.nunique().values
})
display(column_info)
```

### Statistical Summary

```python
# Statistical summary of numerical columns
print("=" * 60)
print("NUMERICAL COLUMNS SUMMARY")
print("=" * 60)
display(df.describe())

# Statistical summary of categorical columns 
print("\n" + "=" * 60)
print("CATEGORICAL COLUMNS SUMMARY")
print("=" * 60)

# Updated to explicitly include string columns
categorical_cols = df.select_dtypes(include=['object', 'string']).columns
for col in categorical_cols:
    print(f"\n{col} value counts:")
    print(df[col].value_counts().head(10))
```

### Target Variable Analysis

```python
# Check class imbalance (critical for fraud detection)
print("=" * 60)
print("TARGET VARIABLE ANALYSIS")
print("=" * 60)

fraud_counts = df['isFraud'].value_counts()
fraud_percentages = df['isFraud'].value_counts(normalize=True) * 100

fraud_summary = pd.DataFrame({
    'Count': fraud_counts,
    'Percentage': fraud_percentages
})
display(fraud_summary)

# Visualize imbalance 
print(f"\nTotal transactions: {len(df):,}")
print(f"Non-fraudulent (0): {fraud_counts[0]:,} ({fraud_percentages[0]:.4f}%)")
print(f"Fraudulent (1): {fraud_counts[1]:,} ({fraud_percentages[1]:.4f}%)")
print(f"\nIMBALANCE RATIO: 1:{fraud_counts[0]/fraud_counts[1]:.2f}")

# Check flagged fraud
print("\n" + "=" * 60)
print("FLAGGED FRAUD ANALYSIS")
print("=" * 60)
flagged_counts = df['isFlaggedFraud'].value_counts()
print("Flagged fraud distribution:")
print(f"  Not Flagged (0): {flagged_counts.get(0, 0):,}")
print(f"  Flagged (1): {flagged_counts.get(1, 0):,}")

# How many flagged frauds are actually fraud?
if flagged_counts.get(1, 0) > 0:
    flagged_and_fraud = df[(df['isFlaggedFraud'] == 1) & (df['isFraud'] == 1)].shape[0]
    print(f"\nFlagged transactions that are actually fraud: {flagged_and_fraud}")
    print(f"Flagged fraud accuracy: {flagged_and_fraud/flagged_counts[1]*100:.2f}%")
```

```python
# Fraud by transaction type
print("=" * 60)
print("FRAUD BY TRANSACTION TYPE")
print("=" * 60)

fraud_by_type = df[df['isFraud'] == 1]['type'].value_counts()
total_by_type = df['type'].value_counts()

fraud_type_summary = pd.DataFrame({
    'Total Transactions': total_by_type,
    'Fraudulent Transactions': fraud_by_type,
    'Fraud Rate (%)': (fraud_by_type / total_by_type * 100).fillna(0)
})
display(fraud_type_summary)

# Which transaction types have fraud?
print("\nTransaction types with fraud:")
for type_name in fraud_by_type.index:
    print(f"  {type_name}: {fraud_by_type[type_name]:,} frauds ({fraud_by_type[type_name]/fraud_counts[1]*100:.1f}% of all frauds)")
```

### Amount Analysis

```python
# Amount analysis - Compare fraud vs legitimate
print("=" * 60)
print("AMOUNT ANALYSIS: FRAUD vs LEGITIMATE")
print("=" * 60)

fraud_amounts = df[df['isFraud'] == 1]['amount']
legit_amounts = df[df['isFraud'] == 0]['amount']

amount_comparison = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Std Dev', 'Min', 'Max', '75th Percentile', '95th Percentile'],
    'Legitimate': [
        f"{legit_amounts.mean():,.2f}",
        f"{legit_amounts.median():,.2f}",
        f"{legit_amounts.std():,.2f}",
        f"{legit_amounts.min():,.2f}",
        f"{legit_amounts.max():,.2f}",
        f"{legit_amounts.quantile(0.75):,.2f}",
        f"{legit_amounts.quantile(0.95):,.2f}"
    ],
    'Fraud': [
        f"{fraud_amounts.mean():,.2f}",
        f"{fraud_amounts.median():,.2f}",
        f"{fraud_amounts.std():,.2f}",
        f"{fraud_amounts.min():,.2f}",
        f"{fraud_amounts.max():,.2f}",
        f"{fraud_amounts.quantile(0.75):,.2f}",
        f"{fraud_amounts.quantile(0.95):,.2f}"
    ]
})
display(amount_comparison)

# Visual insight
print(f"\nInsight: Fraud transactions have {fraud_amounts.mean()/legit_amounts.mean():.1f}x higher average amount than legitimate transactions")
```

### Temporal Analysis

```python
# Temporal analysis - When does fraud happen?
print("=" * 60)
print("TEMPORAL FRAUD PATTERNS")
print("=" * 60)

# Fraud concentration over time
fraud_by_step = df[df['isFraud'] == 1].groupby('step').size()
total_by_step = df.groupby('step').size()

# Find peak fraud hours
print("Top 10 hours with highest fraud activity:")
top_fraud_hours = fraud_by_step.sort_values(ascending=False).head(10)
for hour, count in top_fraud_hours.items():
    total_in_hour = total_by_step[hour]
    fraud_rate = (count / total_in_hour * 100)
    print(f"  Hour {hour}: {count} frauds out of {total_in_hour:,} transactions ({fraud_rate:.2f}% fraud rate)")

# Check if fraud is concentrated in specific time periods
print("\nFraud distribution:")
print(f"  Hours with NO fraud: {len(total_by_step) - len(fraud_by_step)} out of {len(total_by_step)} total hours")
print(f"  Average frauds per hour (when fraud occurs): {fraud_by_step.mean():.2f}")
```

### Customer Analysis

```python
# Customer behavior analysis
print("=" * 60)
print("CUSTOMER BEHAVIOR ANALYSIS")
print("=" * 60)

# Customers involved in fraud
fraud_customers = df[df['isFraud'] == 1]['nameOrig'].unique()
legit_customers = df[df['isFraud'] == 0]['nameOrig'].unique()

print(f"Unique customers (senders): {df['nameOrig'].nunique():,}")
print(f"Customers who committed fraud: {len(fraud_customers):,}")
print(f"Percentage of customers involved in fraud: {len(fraud_customers)/df['nameOrig'].nunique()*100:.4f}%")

# Multiple frauds per customer?
fraud_counts_per_customer = df[df['isFraud'] == 1].groupby('nameOrig').size()
print("\nFraud distribution per customer:")
print(f"  Customers with 1 fraud: {(fraud_counts_per_customer == 1).sum()}")
print(f"  Customers with 2+ frauds: {(fraud_counts_per_customer >= 2).sum()}")
print(f"  Max frauds by one customer: {fraud_counts_per_customer.max()}")

# Check if fraud customers also have legitimate transactions
fraud_customers_with_legit = 0
for customer in fraud_customers[:10]:  # Sample check
    if len(df[(df['nameOrig'] == customer) & (df['isFraud'] == 0)]) > 0:
        fraud_customers_with_legit += 1
print(f"\nSample insight: {fraud_customers_with_legit}/10 fraud customers also made legitimate transactions")
```

### Destination Analysis

```python
# Destination analysis for fraud
print("=" * 60)
print("DESTINATION ANALYSIS FOR FRAUD")
print("=" * 60)

# Check destination types (C for customer, M for merchant)
df['dest_type'] = df['nameDest'].str[0]
fraud_by_dest = df[df['isFraud'] == 1]['dest_type'].value_counts()
total_by_dest = df['dest_type'].value_counts()

dest_summary = pd.DataFrame({
    'Total Transactions': total_by_dest,
    'Fraudulent Transactions': fraud_by_dest,
    'Fraud Rate (%)': (fraud_by_dest / total_by_dest * 100).fillna(0)
})
display(dest_summary)

print("\nKey Finding: Fraud goes to CUSTOMER accounts (C), not merchants (M)")
print(f"   {fraud_by_dest.get('C', 0):,} frauds to customers vs {fraud_by_dest.get('M', 0):,} to merchants")
```

### Balance Fields Analysis

```python
# Analyze the relationship between fraudsters and their destinations
print("=" * 60)
print("FRAUDSTER DESTINATION ANALYSIS")
print("=" * 60)

# Get fraud transactions
fraud_df = df[df['isFraud'] == 1]

# Check if destinations are unique per fraudster
unique_fraud_destinations = fraud_df['nameDest'].nunique()
print(f"Unique destinations for fraud: {unique_fraud_destinations:,}")
print(f"Fraudsters: {len(fraud_df):,}")
print(f"Ratio (destinations/fraudsters): {unique_fraud_destinations/len(fraud_df):.2f}")

# Check if destinations receive multiple frauds
dest_fraud_counts = fraud_df['nameDest'].value_counts()
print("\nDestinations receiving multiple frauds:")
print(f"  Destinations with 1 fraud: {(dest_fraud_counts == 1).sum()}")
print(f"  Destinations with 2+ frauds: {(dest_fraud_counts >= 2).sum()}")
if (dest_fraud_counts >= 2).sum() > 0:
    print(f"  Max frauds to one destination: {dest_fraud_counts.max()}")

# Are destinations also regular customers?
fraud_destinations = set(fraud_df['nameDest'].unique())
regular_customers = set(df[df['isFraud'] == 0]['nameOrig'].unique())
overlap = fraud_destinations.intersection(regular_customers)
print(f"\nFraud destinations that are also regular customers: {len(overlap)}")
```

### Summary Report

```python
# Generate a comprehensive summary
print("=" * 60)
print("DATASET EXPLORATION SUMMARY")
print("=" * 60)

import json
from datetime import datetime

# Create a summary dictionary of all findings
analysis_summary = {
    'dataset_info': {
        'total_transactions': int(len(df)),
        'total_columns': int(len(df.columns)),
        'memory_mb': float(df.memory_usage(deep=True).sum() / (1024 * 1024)),
        'time_range_hours': f"{df['step'].min()}-{df['step'].max()}",
        'time_range_days': float((df['step'].max() - df['step'].min()) / 24)
    },
    'fraud_statistics': {
        'fraud_count': int(fraud_counts[1]),
        'legit_count': int(fraud_counts[0]),
        'fraud_percentage': float(fraud_percentages[1]),
        'imbalance_ratio': float(fraud_counts[0] / fraud_counts[1])
    },
    'flagged_fraud': {
        'flagged_count': int(flagged_counts.get(1, 0)),
        'flagged_accuracy': float(100.0),
        'detection_rate': float((flagged_and_fraud / fraud_counts[1] * 100)) if flagged_counts.get(1, 0) > 0 else 0
    },
    'transaction_types': {
        'types_with_fraud': ['CASH_OUT', 'TRANSFER'],
        'cash_out_frauds': int(fraud_by_type.get('CASH_OUT', 0)),
        'transfer_frauds': int(fraud_by_type.get('TRANSFER', 0)),
        'cash_out_percentage': float(fraud_by_type.get('CASH_OUT', 0) / fraud_counts[1] * 100),
        'transfer_percentage': float(fraud_by_type.get('TRANSFER', 0) / fraud_counts[1] * 100)
    },
    'amount_analysis': {
        'avg_legit_amount': float(legit_amounts.mean()),
        'avg_fraud_amount': float(fraud_amounts.mean()),
        'median_legit_amount': float(legit_amounts.median()),
        'median_fraud_amount': float(fraud_amounts.median()),
        'amount_multiplier': float(fraud_amounts.mean() / legit_amounts.mean())
    },
    'customer_analysis': {
        'unique_customers': int(df['nameOrig'].nunique()),
        'fraud_customers': int(len(fraud_customers)),
        'fraud_customer_percentage': float(len(fraud_customers) / df['nameOrig'].nunique() * 100),
        'max_frauds_per_customer': int(fraud_counts_per_customer.max()) if len(fraud_counts_per_customer) > 0 else 0
    },
    'destination_analysis': {
        'fraud_to_customers': int(fraud_by_dest.get('C', 0)),
        'fraud_to_merchants': int(fraud_by_dest.get('M', 0)),
        'unique_fraud_destinations': int(unique_fraud_destinations),
        'destinations_with_multiple_frauds': int((dest_fraud_counts >= 2).sum())
    },
    'temporal_analysis': {
        'hours_with_100_percent_fraud': ['523', '501', '387', '425', '730'],
        'hours_with_no_fraud': 2,
        'avg_frauds_per_hour': float(fraud_by_step.mean()) if len(fraud_by_step) > 0 else 0
    }
}

# Save as JSON for easy reference
with open('../artifacts/analysis_summary.json', 'w') as f:
    json.dump(analysis_summary, f, indent=2)

print("Analysis summary saved to: artifacts/analysis_summary.json")

# Create a README with all key findings and next steps
readme_content = f"""
# Financial Fraud Detection Project - Data Analysis

## Dataset Overview
- **Total Transactions:** {len(df):,}
- **Time Period:** {df['step'].min()}-{df['step'].max()} hours ({analysis_summary['dataset_info']['time_range_days']:.1f} days)
- **File Size:** {analysis_summary['dataset_info']['memory_mb']:.2f} MB

## Critical Findings
1. **Fraud Rate:** {fraud_percentages[1]:.4f}% ({fraud_counts[1]:,} frauds)
2. **Imbalance Ratio:** 1:{analysis_summary['fraud_statistics']['imbalance_ratio']:.2f}
3. **Fraud Only In:** CASH_OUT ({fraud_by_type.get('CASH_OUT', 0):,}) and TRANSFER ({fraud_by_type.get('TRANSFER', 0):,})
4. **Amount Pattern:** Fraud amounts are {analysis_summary['amount_analysis']['amount_multiplier']:.1f}x higher
5. **Destination Pattern:** 100% of frauds go to customer accounts (never merchants)
6. **User Pattern:** Each fraudster has exactly 1 transaction (no history)

## Files Created
- `analysis_summary.json` - Complete analysis results
- Raw data saved to: `artifacts/raw/Synthetic_Financial_datasets_log.csv`

**Important Note:** Since fraudsters have no transaction history, use GLOBAL time-based sequences, not per-user sequences!
"""

with open('../artifacts/DATA_ANALYSIS_COMPLETE.md', 'w') as f:
    f.write(readme_content)

print("README saved to: artifacts/DATA_ANALYSIS_COMPLETE.md")
```
